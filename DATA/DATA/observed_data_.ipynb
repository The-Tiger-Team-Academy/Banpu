```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the provided CSV file
file_path = 'df_jepx_spotsum.csv'
data = pd.read_csv(file_path)

# Preparing the data
data['DATETIME'] = pd.to_datetime(data['DATETIME'])
data.set_index('DATETIME', inplace=True)

# Calculate EMAs
data['EMA_12'] = data['HOKKAIDO'].ewm(span=50, adjust=False).mean()
data['EMA_26'] = data['HOKKAIDO'].ewm(span=100, adjust=False).mean()

# Generate signals
data['signal'] = 0
previous_signal = 0

for i in range(1, len(data)):
    if data['EMA_12'].iloc[i] > data['EMA_26'].iloc[i] and previous_signal != 1:
        data['signal'].iloc[i] = 1  # Buy signal
        previous_signal = 1
    elif data['EMA_26'].iloc[i] > data['EMA_12'].iloc[i] and previous_signal != -1:
        data['signal'].iloc[i] = -1  # Sell signal
        previous_signal = -1

# Initialize balance and position tracking
initial_balance = 10000  # Starting balance in dollars
balance = initial_balance
position = 0  # Current position
data['balance'] = initial_balance

# Backtest the strategy
for i in range(1, len(data)):
    if data['signal'].iloc[i] == 1 and position <= 0:
        # Buy
        position += balance / data['HOKKAIDO'].iloc[i]
        balance = 0
    elif data['signal'].iloc[i] == -1 and position > 0:
        # Sell
        balance += position * data['HOKKAIDO'].iloc[i]
        position = 0
    data['balance'].iloc[i] = balance + position * data['HOKKAIDO'].iloc[i]

# Plot Buy/Sell Signals on Price
plt.figure(figsize=(12, 6))
plt.plot(data['HOKKAIDO'], label='HOKKAIDO Price', alpha=0.5)
plt.plot(data['EMA_12'], label='EMA 12', alpha=0.5)
plt.plot(data['EMA_26'], label='EMA 26', alpha=0.5)
plt.scatter(data.loc[data['signal'] == 1].index, data.loc[data['signal'] == 1]['HOKKAIDO'], label='Buy Signal', marker='^', color='green')
plt.scatter(data.loc[data['signal'] == -1].index, data.loc[data['signal'] == -1]['HOKKAIDO'], label='Sell Signal', marker='v', color='red')
plt.title('Buy and Sell Signals on HOKKAIDO Price')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()

# Plot Total Balance Over Time
plt.figure(figsize=(12, 6))
plt.plot(data['balance'], label='Total Balance')
plt.title('Total Balance Over Time')
plt.xlabel('Time')
plt.ylabel('Balance')
plt.legend()

plt.show()
```
